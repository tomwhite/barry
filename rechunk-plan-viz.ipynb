```python
import cubed
import cubed as xp
from IPython.display import display
```

```python
shape = (350640, 721 * 1440)
source_chunks = (31, 721 * 1440)
target_chunks = (350640, 10 * 10)
```

```python
a = xp.empty(shape, dtype=xp.float32, chunks=source_chunks)
b = a.rechunk(target_chunks)
```

```python
cubed.visualize(b)
```

```python
# Show arrays in plan
for n, d in b.plan.dag.nodes(data=True):
    if n.startswith("array-"):
        target = d["target"]
        if not isinstance(target, cubed.storage.zarr.LazyZarrArray):
            continue
        print(n)
        arr = xp.empty(target.shape, dtype=target.dtype, chunks=target.chunks)
        display(arr)
```
